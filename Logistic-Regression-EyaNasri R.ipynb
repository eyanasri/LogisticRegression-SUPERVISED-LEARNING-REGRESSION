# Regression Logistique avec R

### Les données traitées dans ce TP décrivent des individus ayant des revenus >50K ou <50K. Le but est d’utiliser les variables explicatives de ces données afin de prédire la variable expliquée binaire (le label) qui est le revenu de chaque personne.

```r
dataset=read.csv("census_income.csv")
dataset
```

```r
summary(dataset)
```

```r
# La fonction de la regression logistique
? glm
```

```r
#GLM avec R s'utilise globalement comme la fonction lm(), 
#L'option 'family=binomial' pour le lien LOGIT

modele <- glm(income_level ~ capital_gain+capital_loss+hours_per_week, data=dataset, family=binomial)
modele
summary(modele)
pred.proba <- predict(modele, newdata= dataset, type="response")
pred.proba
```

```r
#Appliquer la fonction de décision
p <- factor(ifelse(pred.proba > 0.5, "presence", "absence")) 
p
```

```r
#Matrice de confusion
mc <- table( dataset$income_level, p ) 
mc
```
